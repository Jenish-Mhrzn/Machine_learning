```python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/College/Video_Games_Sales_as_at_22_Dec_2016.csv")
df.head(5)
```

```python
#showing null values datas
df[df.isnull().any(axis=1)]
df.head(5)
```

```python
#df.dropna() will remove epmty cells but will not chnage the dataframe
new_df=df.dropna()
new_df
```

```python
df.head(5)
```

```python
#another way of handling the null values is by replacing the null values with some other value
new1_df=df.fillna(130)
new1_df
```

```python
df.head(5)
```

```python
#filling the null values of colums as random value
df["Critic_Count"].fillna(62, inplace=True)
df
```

```python
#fill null values with mean, median or mode
x=df["Critic_Score"].mean()
df["Critic_Score"]=df["Critic_Score"].fillna(x)
y=df["Critic_Count"].mean()
df["Critic_Count"]=df["Critic_Count"].fillna(y)
z=df["User_Score"].mean()
df["User_Score"]=df["User_Score"].fillna(z)
a=df["User_Count"].mean()
df["User_Count"]=df["User_Count"].fillna(a)
df
```

```python
df["Developer"]=df["Developer"].fillna("Nintendo")
df
```

```python
df["Rating"]=df["Rating"].fillna("E")
df
```
